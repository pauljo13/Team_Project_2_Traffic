```python
def image_list(data_list):
    names = [i[0:7] for i in data_list]
    counter = {k: v for k, v in Counter(names).items() if v >= 5}
    new_list = [j for j in data_list if j[0:7] in counter]
    return new_list, list(set([i[0:7] for i in new_list]))

def cctv_one_list(data_list, cctv):
    return sorted([item for item in data_list if cctv in item])

def create_gif_from_images(image_list, image_path, output_path, fps=0.5):
    full_image_paths = [os.path.join(image_path, image_name) for image_name in image_list]
    images = [Image.open(image_path) for image_path in full_image_paths]
    duration = 1000 / fps
    gif_filename = os.path.join(output_path, f"{image_list[0][0:7]}.gif")
    imageio.mimsave(gif_filename, images, duration=duration)
    print(f"GIF saved at: {gif_filename}")

def EDA_image_gif(filenames, input_url, output_url):
    data_list, cctv_names = image_list(filenames)
    print(len(cctv_names))
    count = 0
    for i in cctv_names:
        single_cctv_images = cctv_one_list(data_list, i)
        create_gif_from_images(single_cctv_images, input_url, output_url)
        count += 1
        print(f'{count} cycle complete...')
    print('완료했습니다!!!')
```

```python
import torch
import imageio
import numpy as np

# YOLOv5 모델 불러오기
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

def detect_objects(frame):
    # frame을 NumPy 배열로 변환하고, 객체 탐지 실행
    results = model(np.array(frame))
    # 탐지된 객체에 바운딩 박스 그리기
    rendered_img = results.render()[0]
    # OpenCV에서는 BGR 포맷을 사용하므로 RGB로 변환
    return rendered_img

# GIF 파일 불러오기
gif_path = '/Users/bumsoojoe/Desktop/팀프로젝트2_교통법규/database/gif/C000003.gif'
frames = [frame for frame in imageio.get_reader(gif_path)]

# 각 프레임에 대해 객체 탐지 실행
detected_frames = [detect_objects(frame) for frame in frames]

# 수정된 프레임으로 새로운 GIF 생성
# fps=24를 duration=1000/24로 대체
imageio.mimsave('output.gif', detected_frames, duration=1000/24)
```

```python
import os
import json
import cv2
import imageio
from tqdm import tqdm

# gif를 frame으로 변환
def gif_to_frames(gif_path, save_folder):
    gif = imageio.get_reader(gif_path)
    for i, frame in enumerate(gif):
        imageio.imwrite(f'{save_folder}/frame_{i:04}.jpg', frame)

def convert_json_to_yolo(json_path, image_path, yolo_save_folder):
    with open(json_path, 'r') as jf:
        data = json.load(jf)

    img = cv2.imread(image_path)
    height, width = img.shape[:2]

    annotations = []

    for ann in data["annotations"]["PM"]:
        if ann["shape_type"] == "bbox":
            x1, y1 = int(ann["points"][0]), int(ann["points"][1])
            x2, y2 = int(ann["points"][2]), int(ann["points"][3])
            
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

            x_center = (x1 + x2) / 2 / width
            y_center = (y1 + y2) / 2 / height
            w = (x2 - x1) / width
            h = (y2 - y1) / height
            annotations.append(f"0 {x_center} {y_center} {w} {h}")

    cv2.imwrite(image_path, img)

    save_name = os.path.basename(image_path).replace(".jpg", ".txt")
    save_path = os.path.join(yolo_save_folder, save_name)

    with open(save_path, 'w') as f:
        f.write('\n'.join(annotations))

gif_file_path = '/Users/bumsoojoe/Desktop/팀프로젝트2_교통법규/database/gif/C000003.gif'
frame_save_folder = 'frames_from_gif'
os.makedirs(frame_save_folder, exist_ok=True)
gif_to_frames(gif_file_path, frame_save_folder)

yolo_annotations_folder = 'yolo_annotations'
os.makedirs(yolo_annotations_folder, exist_ok=True)

json_files = []
for i in single_cctv_images:
    json_files.append(f'/Users/bumsoojoe/Desktop/팀프로젝트2_교통법규/120.개인형 이동장치 안전 데이터/01.데이터/2.Validation/라벨링데이터/VL1/CCTV/야간/맑음/실증/{i}')

for i, json_file in tqdm(enumerate(json_files)):
    json_path = os.path.join('path_to_json_folder', json_file)
    image_path = os.path.join(frame_save_folder, f"frame_{i:04}.jpg")
    convert_json_to_yolo(json_path, image_path, yolo_annotations_folder)

# # 임시로 저장한 프레임 이미지들을 제거
# for file_name in os.listdir(frame_save_folder):
#     os.remove(os.path.join(frame_save_folder, file_name))

# os.rmdir(frame_save_folder)  # 폴더 자체도 제거
```
